DATA VISUALIZATION WITH MATPLOT  

```python
#importing the libraries
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing the dataset
mtcars = pd.read_csv("mtcars.csv")
```

```python
mtcars.head()
```

```python
mtcars.info()
```

```python
mtcars.columns
```

```python
mtcars.shape
```

```python
#TABLE
pd.crosstab(mtcars.mpg,mtcars.cyl)
```

```python
pd.crosstab(mtcars.mpg,mtcars.hp)
```

```python
pd.crosstab(mtcars.mpg,mtcars.disp)
```

```python
#crosstable with bar plot
pd.crosstab(mtcars.gear,mtcars.cyl).plot(kind='bar')
```

```python
pd.crosstab(mtcars.hp,mtcars.cyl).plot(kind='bar')
```

```python
#scatter plt
plt.xlabel('Cylinder')
plt.ylabel('MPG')
plt.title('Cylinder vs Mpg')
plt.suptitle('COMPAIRSON',size=25,y=1.050)
plt.scatter(mtcars.cyl,mtcars.mpg)
```

```python
plt.scatter(mtcars.cyl,mtcars.gear)
plt.title('Cylinder vs gear')
plt.xlabel('Cylinder')
plt.ylabel('gear')
plt.suptitle('COMPAIRSON',size = 25,y=1.085)
```

```python
plt.scatter(mtcars.mpg,mtcars.qsec)
plt.xlabel('Mpg')
plt.ylabel('Qsec')
plt.title("Mpg vs Qsec")
plt.suptitle("COMPAIRSON",size=25,y=1.080)
```

```python
#pie plot
mtcars['gear'].value_counts().plot(kind='pie')
```

```python
mtcars['cyl'].value_counts().plot(kind='pie')
```

```python
#Histogram
plt.hist(mtcars['mpg'],facecolor="purple",edgecolor="white",bins=5)
```

```python
plt.hist(mtcars['cyl'],facecolor='green',edgecolor='white',bins=5)
```

```python
#boxplot
plt.boxplot(mtcars['qsec'],vert=True );plt.title('BOXPLOT');plt.ylabel('Range')
```

```python
#violin plot
plt.violinplot(mtcars['mpg'])
```
