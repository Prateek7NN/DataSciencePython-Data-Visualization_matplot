DATA VISUALIZATION WITH SEABORN

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the dataset
tips=pd.read_csv("tips.csv")
```

```python
tips.head()
```

```python
tips.info()
```

Univariate -"one variable" data visualization

```python
#strip plot
sns.stripplot(y='tip',data=tips,jitter=False)
plt.ylabel('tips($)')
plt.show()

```

```python
sns.stripplot(y='size',data=tips,jitter=True)
plt.ylabel("size")
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=True)
plt.show()
```

```python
sns.stripplot(x='day',y='size',data=tips,size=10.5,jitter=True)
plt.ylabel('size')
plt.show()
```

```python
sns.stripplot(x='size',y='tip',data=tips,hue='sex',size=4,jitter=True)
plt.show()
```

#swarm plot

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.show()
```

```python
sns.swarmplot(x='tip',y='day',hue='sex',orient='h',data=tips)
plt.show()
```

```python
sns.swarmplot(x='day',y='time',hue='sex',data=tips,orient='h' )
plt.show()
```

```python
tips.columns
```

```python
#Box Plot
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
```

```python
#violin plot
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.show()
```

```python
#combining plot
sns.violinplot(x='day',y='tip', data=tips,inner=None,color='lightgrey')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.show()
```

BIVARIATE data visualization

```python
#joint plot
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#density plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

```python
#pairplot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```
